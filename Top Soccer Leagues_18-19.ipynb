```python
#Remove Unnecessary Columns From English League and sort data based on best scorer
new_english = english.filter(['Rk','Player','Nation','Pos','Squad','Age','Born','MP','Starts','Min','Gls','Ast','PK','PKatt','CrdY','CrdR'], axis=1)
league = "EPL"
new_english.insert(0, 'League', league)
new_english.reset_index(drop=True, inplace=True)
new_english.index += 1 
new_english.index
new_english.head()
```

```python
new_english.describe()
```

```python
# EPL Top 10 Scorers
new_english = new_english.sort_values('Gls',ascending=False)
new_english.reset_index(drop=True, inplace=True)
new_english.index += 1 
new_english.index
new_english.head(10)
```

```python
# EPL Top 20 Players who played the highest number of minutes not matches
new_english = new_english.sort_values('Min',ascending=False)
new_english.reset_index(drop=True, inplace=True)
new_english.index += 1 
new_english.index
new_english.head(20)
```

```python
#Remove Unnecessary Columns From German League and sort data based on best scorer
new_german = german.filter(['Rk','Player','Nation','Pos','Squad','Age','Born','MP','Starts','Min','Gls','Ast','PK','PKatt','CrdY','CrdR'], axis=1)
league = "Bundesliga"
new_german.insert(0, 'League', league)
new_german.reset_index(drop=True, inplace=True)
new_german.index += 1 
new_german.index
new_german.head()
```

```python
# Bundesliga Top 20 Players who played the highest number of minutes not matches
new_german = new_german.sort_values('Min',ascending=False)
new_german.reset_index(drop=True, inplace=True)
new_german.index += 1 
new_german.index
new_german.head(20)
```

```python
#Remove Unnecessary Columns From Spanish League and sort data based on best scorer
new_spanish = spanish.filter(['Rk','Player','Nation','Pos','Squad','Age','Born','MP','Starts','Min','Gls','Ast','PK','PKatt','CrdY','CrdR'], axis=1)
league = "La Liga"
new_spanish.insert(0, 'League', league)
new_spanish = new_spanish.sort_values('Gls',ascending=False)
new_spanish.reset_index(drop=True, inplace=True)
new_spanish.index += 1 
new_spanish.index
new_spanish.head()
```

```python
#Merge all dataframes
#new_german.merge(new_english, on='League')
data_frames = [new_german, new_spanish, new_english]
top_leagues=pd.concat(data_frames)
top_leagues.reset_index(drop=True, inplace=True)
top_leagues.index += 1 
top_leagues.index
top_leagues.head()
```

```python
#Save Overall File As CSV
top_leagues.to_csv('Top_Leagues.csv')
```

```python
# Overall Top 10 Scorers
top_leagues = top_leagues.sort_values('Gls',ascending=False)
top_leagues.reset_index(drop=True, inplace=True)
top_leagues.index += 1 
top_leagues.index
top_leagues.head(10)
```

```python
# La Liga Top 20 Players who played the highest number of minutes not matches
top_leagues = top_leagues.sort_values('Min',ascending=False)
top_leagues.reset_index(drop=True, inplace=True)
top_leagues.index += 1 
top_leagues.index
top_leagues.head(20)
```

```python
#total_goal_per_team

goals = top_leagues.groupby('Squad') 
overall_info = goals[['MP','Min','Gls','Ast','PK','CrdY','CrdR']]
overall_info.sum()
```
